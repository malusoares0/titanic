### Juntando gender_submission com test

<br>

```python
#Para juntar gender_submission e test, utilizando a coluna PassengerId (em comum) como chave
df_test_gender = pd.merge(gender_submission, test, on="PassengerId", how="inner")  

#checar a qntd de linhas
```

```python
#verificando
df_test_gender

#deu certo
#linhas ok
```

<br>

### Mudando o nome de gender_submission porque não há relação entre o nome e o conteúdo

<br>

```python
df_test_death = df_test_gender.copy()
```

```python
#verificando a mudança
df_test_death

#deu certo
```

<br>

### Concatenando df_test_death e train, para ter um único dataframe

<br>

```python
titanic = pd.concat([train, df_test_death], axis=0, sort=False).reset_index(drop=True)
```

```python
#verificando
titanic

#deu certo
```

```python
#fazendo uma cópia do dataframe, para não alterar os dados do df original
df_titanic = titanic.copy()
```

```python
#verificando
df_titanic.sample(15)

#deu certo
```

```python
df_titanic.info()
```

<br>

### Colocando todas as letras dos nomes das colunas em minúsculo

<br>

```python
df_titanic.columns
```

```python
#List comprehension para passar todas as letras das colunas para minúsculas
df_titanic.columns = [col.lower() for col in df_titanic.columns] 

```

```python
#verificando
df_titanic

#deu certo
```
